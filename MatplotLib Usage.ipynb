# LINE Plot

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
y=2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line PLOT")
plt.xlabel("X Asix")
plt.ylabel("Y Asix")
plt.show()
```

```python
plt.plot(x,y,color="purple",linestyle="-.",linewidth=3)
plt.show()
```

```python
x = np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color="purple",linestyle="-.",linewidth=3)
plt.plot(x,y2,color="red",linestyle=":",linewidth=2)
plt.show()
```

```python
plt.plot(x,y1,color="purple",linestyle="-.",linewidth=3)
plt.plot(x,y2,color="red",linestyle=":",linewidth=2)
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color="purple",linestyle="-.",linewidth=3)
plt.subplot(1,2,2)
plt.plot(x,y2,color="red",linestyle=":",linewidth=2)
plt.show()
```

# Bar PLOT

```python
student={"Bob":87,"LK":99,"Matt":54,"Alex":37}

names=list(student.keys())
values = list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

# Horizontal Graph

```python
plt.barh(names,values,color='g')
plt.title("BAR -H- Plot")
plt.xlabel("Names")
plt.ylabel("Values")
plt.grid()
plt.show()
```

# Scatter Plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
```

```python
plt.scatter(x,a)
plt.show()
```

### Changing Aesthetics

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
```

```python
plt.scatter(x,a,marker='*',c='grey',s=100)
plt.show()
```

### Adding 2 matrix in same plot

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,6,9,1,4,5,3]
```

```python
plt.scatter(x,a,marker='*',c='grey',s=100)
plt.scatter(x,b,marker='*',c='orange',s=300)
plt.show()
```

### Adding Sub Plots

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='red',s=100)
plt.subplot(1,2,2)
plt.scatter(x,b,marker='.',c='blue',s=300)
plt.show()
```

# BOX Plot

```python
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data= list([one,two,three])
```

```python
plt.boxplot(data)
plt.show()
```

# Violin Plot

```python
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]

data= list([one,two,three])

plt.violinplot(data,showmedians=True)
plt.show()
```

```python
one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]

data= list([one,two,])

plt.violinplot(data)
plt.show()
```

# Pie Chart

```python
fruit=["Apple","Mango","Orange","Guavan","Banana"]
quantity=[67,34,100,46,81]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

### Changing Aesthetics

```python
plt.pie(quantity,labels=fruit,autopct='%0.2f%%',colors=['yellow','blue','black','grey','pink'])
plt.show()
```

# DoughNut Chart

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
